# SafePrice Simulator

### Import Libraries

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Reading Raw Datasets

```python
root_path = os.path.dirname(os.getcwd())

raw_data_path = os.path.join(root_path, "datasets/raw")
raw_datasets = os.listdir(raw_data_path)

print(raw_datasets)
```

#### Read Retail Dateset

```python
online_retail_df_2009_2010 = pd.read_excel(os.path.join(raw_data_path, raw_datasets[0]), sheet_name='year_2009_2010')

online_retail_df_2009_2010.head()
```

```python
online_retail_df_2010_2011 = pd.read_excel(os.path.join(raw_data_path, raw_datasets[0]), sheet_name='year_2010_2011')

online_retail_df_2010_2011.head()
```

```python
retail_df = pd.concat([online_retail_df_2009_2010, online_retail_df_2010_2011], ignore_index=True)
retail_df.tail()
```

```python
print(len(online_retail_df_2009_2010))
print(len(online_retail_df_2010_2011))
print(len(retail_df))
```

#### Read Retail Prices Index Dateset

```python
retail_price_index_df = pd.read_excel(os.path.join(raw_data_path, raw_datasets[1])).iloc[972:1008]
retail_price_index_df.head()
```

```python
retail_price_index_df.rename(columns={
    'Title': 'Date',
    'Retail Prices Index: Long run series: 1800 to 2024: Jan 1974=100': 'Retail Price Index'
}, inplace=True)
```
